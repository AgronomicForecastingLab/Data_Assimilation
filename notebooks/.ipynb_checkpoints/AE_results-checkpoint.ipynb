### Comparing AE results with variable latent dimension


```python
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_dir = "/home/jfm1118/results/"
test_size = 327
train_size = 659
```

```python
results_fps = os.listdir(results_dir)
sorted(results_fps)
```

```python
def extract_test_train(ls, dim):
    """Helper function to extract the test and train fp of a given 
    dimension dim, from list ls"""
    test, train = None, None
    for fp in ls:
        text = fp.replace("epoch120.txt", "")
        split = text.split("_dim" + str(dim) + "_")
        if len(split) == 2:
            first, _ = split
            if first == "test":
                test = fp
            elif first == "train":
                train = fp
            else:
                raise ValueError("Should have a filename starting with \'train\' or \'test\'")
    if test and train:
        return train, test
    else:
        raise ValueError("No test or train fp present in ls")
def plot_dim(dim, ax):
    train, test = extract_test_train(results_fps, dim)
    
    with open(results_dir + train, 'rb') as fp:
        train = pickle.load(fp)
    with open(results_dir + test, 'rb') as fp:
        test = pickle.load(fp)
    epoch, data = list(zip(*test))
    data = np.array(data) / test_size
    axes = ax.plot(epoch, data, label = "test")
    
    epoch, data = list(zip(*train))
    data =  np.array(data) / train_size
    
    ax.plot(epoch, data, label = "train")
    ax.legend()
    ax.set_xlim([1, 50])
    ax.set_ylim([0, 800])
    return axes
extract_test_train(results_fps, 1)
```

```python
latent_dims = [1, 2, 4, 10, 20, 40, 75, 100, 150, 200]
rows = 5
cols = 2
fig, ax =plt.subplots(rows, cols, sharex = True,sharey = True, figsize = (10, 20))

for idx, dim in enumerate(latent_dims):
    pos_x = idx % cols
    pos_y = idx // cols
    axes = ax[pos_y, pos_x]
    plot_dim(dim, axes)
    axes.set_title("Latent dimensions = {}".format(dim))
fig.show()
    
    

```
