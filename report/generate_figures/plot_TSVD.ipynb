```python
def get_num(file, ):
    num = ""
    for idx, c in enumerate(file):
        if c.isnumeric():
            num += c
        else:
            break
    if num is "":
        assert file[:4] == "None"
        idx = 3
        num = None
    else:
        num = int(num)
        
    return num, idx

def get_tsvd_best_df(fp, prov_nobs, prov_modes):
    results = []
    for path, subdir, files in os.walk(fp):
        for file_orig in files:
            file = file_orig
            if file[-4:] != ".csv":
                continue
            file = file.replace("modes", "")
            
            modes, idx = get_num(file)

            file = file[idx:]
            
            if file[0] == "A":
                file = file[3:]
                nobs = "ALL"
            elif file[0] == "_":
                file = file[1:]
                nobs, idx = get_num(file)
            fp = os.path.join(path, file_orig)
            vals = []
            df = pd.read_csv(fp)
            
            if nobs == prov_nobs and modes == prov_modes:
                return df
            
    return None    

def get_tsvd_data(fp, keys, drop_first=False):
    results = []
    for path, subdir, files in os.walk(fp):
        for file_orig in files:
            file = file_orig
            if file[-4:] != ".csv":
                continue
            file = file.replace("modes", "")
            
            modes, idx = get_num(file)

            file = file[idx:]
            
            if file[0] == "A":
                file = file[3:]
                nobs = "ALL"
            elif file[0] == "_":
                file = file[1:]
                nobs, idx = get_num(file)
            fp = os.path.join(path, file_orig)
            vals = []
            for kidx, key_ in enumerate(keys):
                df = pd.read_csv(fp)
                if drop_first[kidx]:
                    df = df.drop(0)
                val = df[key_].mean()
                vals.append(val)
            vals = tuple(vals)
            results.append(vals + (nobs, modes))
    res = pd.DataFrame(results, columns=keys + [ "nobs", "modes"])
    return res      
keys = ["mse_DA", "time"]
drop = [0, 1]

nobsdf = get_tsvd_data(nobsfp, keys, drop)
modesdf = get_tsvd_data(modesfp, keys, drop)
modesdf
```

```python

def plot_val(val, dfs, title, savefp, legend_labels=None, Tucodec_mse=0.0826, Tucodec_time=0.371):
    colours = ["g", "r", "b"]
    
    def sort_vals(df, val, yval):
        df = df.copy()
        x = df[val]
        y = df[yval]
        lists = sorted(zip(*[x, y]))
        new_x, new_y = list(zip(*lists))
        return new_x, new_y
    
    fig, axs = plt.subplots(1, 2,  sharey=False)
    
    if not isinstance(dfs, list):
        dfs = [dfs]
    assert len(colours) >= len(dfs)
    if len(dfs) > 1:
        assert legend_labels is not None
        assert len(dfs) == len(legend_labels)
    
    lines = []
    for idx, df in enumerate(dfs):
        x, y = sort_vals(df, val, "mse_DA")

        #plot left
        ax = axs[0]
        ax.plot(x, y, "-" + colours[idx])
        ax.set_xlabel(title)
        ax.set_ylabel("DA MSE")

        if val == "nobs":
            ax.set_xscale("log", basex=2)
            ax.xaxis.set_major_formatter(ScalarFormatter())
        
        ax = axs[1] 
        x, y = sort_vals(df, val, "time")
        l = ax.plot(x, y, "-" + colours[idx])
        lines.append(l[0])
        ax.set_xlabel(title)
        ax.set_ylabel("Time Taken (s)")
        fig.set_size_inches(15, 5.5)
        if val== "nobs":
            #ax.set_xscale("log", basex=2)
            ax.xaxis.set_major_formatter(ScalarFormatter())
        
        
    
    if len(dfs) > 1:
        ltitle = None
        if val == "nobs":
            ltitle = "Modes"
        elif val == "modes":
            ltitle = "N. Obs."
        axs[0].legend(lines, legend_labels, title=ltitle)
    
    
    #add dotted lines for tucodec values
    x = np.linspace(0.1,1e7,10)
    y = 0 * x + Tucodec_mse
    axs[0].plot(x, y, '--b')
    
    x = np.linspace(-10,1e7,10)
    y = 0 * x + Tucodec_time
    
    axs[1].plot(x, y, '--b', label="Tucodec")
    
    if val== "nobs":
        xmax = 150000
        #xmax = 91 * 85 * 32 TODO - uncomment
        xmin = 0.5
    else:
        xmax = 800
        xmin = -10
    axs[0].set_xlim(xmin, xmax)
    axs[1].set_xlim(xmin, xmax)
    axs[0].set_ylim(0.05, None)

    #add floating tucodec label
    if val== "nobs":
        x_pos0 = 2
        y_pos0 = Tucodec_mse + 0.01
        x_pos1 = 120000
        #x_pos1 = 150000 UNCOMMENT
        
    else:
        x_pos0 = 75
        y_pos0 = Tucodec_mse - 0.013
        x_pos1 = 75
    axs[0].text(x_pos0, y_pos0, 'Tucodec', color = "b")
    axs[1].text(x_pos1, Tucodec_time + 0.05, 'Tucodec', color = "b")

    fig.savefig(savefp)
    
nobs32 = nobsdf[nobsdf["modes"] == 32]
nobsother = nobsdf[nobsdf["modes"] != 32]

legend_labs = ["32", "731"]

plot_val("nobs", [nobs32, nobsother], "Number of Observations", outfp1, legend_labs)

modesAll = modesdf[modesdf["nobs"] != 2475]
modes2475 = modesdf[modesdf["nobs"] == 2475]

legend_labs = ["2475", "247520"]

plot_val("modes", [modes2475, modesAll], "Truncation parameter", outfp2, legend_labs)
```
